```python
import dash
import pandas as pd
import plotly.express as px
from dash import dcc
from dash import html
```

```python
airline_df = pd.read_csv('./airline_data.csv')
airline_df.shape
```

```python
airline_df.columns
```

```python
pd.set_option('display.max_columns', None)
```

```python
airline_df['Flights'].unique()
```

```python
# Pie-chart
fig = px.pie(airline_df,
             values='DistanceGroup',
             names='DistanceGroup',
             title='Distance group proportion by flights')
fig.show()
```

```python
# # create a Dash application:
# app = dash.Dash(__name__)
# 
# # create the layout of the application:
# app.layout = html.Div(children=[
#     html.H1('Airline Dashboard'),
#     html.P('Proportion of distance group (250 miles distance interval group) by flights'),
#     dcc.Graph(figure=fig)
# ])
# 
# # Run the application
# if __name__ == '__main__':
#     print('Running the application')
#     app.run_server()
#create a dash application
app = dash.Dash(__name__)

#create the layout of the application
app.layout = html.Div(
    children=[html.H1('Airline Dashboard', style={'textAlign': 'center', 'color': '#95341f', 'font_size': 40}),
              html.P('Proportion of distance group (500 mile distance interval group) flights'),
              dcc.Graph(figure=fig)
              ])

#run the application
if __name__ == '_main_':
    app.run_server()
```
