```python

df['1980_total'] = df.iloc[:, 4:14].sum(numeric_only=True, axis=1)
df['1990_total'] = df.iloc[:, 14:24].sum(numeric_only=True, axis=1)
df['2000_total'] = df.iloc[:, 24:34].sum(numeric_only=True, axis=1)
df['2010_total'] = df.iloc[:, 34:48].sum(numeric_only=True, axis=1)
```

```python
africa_data = {}
asia_data = {}
europe_data = {}
oceania_data = {}
latin_data = {}
north_data = {}

for index, row in df.iterrows():
    if row['Continent'] == 'Africa':
        africa_data['1980_total'] = africa_data.get('1980_total', 0) + row['1980_total']
        africa_data['1990_total'] = africa_data.get('1990_total', 0) + row['1990_total']
        africa_data['2000_total'] = africa_data.get('2000_total', 0) + row['2000_total']
        africa_data['2010_total'] = africa_data.get('2010_total', 0) + row['2010_total']
    if row['Continent'] == 'Asia':
        asia_data['1980_total'] = asia_data.get('1980_total', 0) + row['1980_total']
        asia_data['1990_total'] = asia_data.get('1990_total', 0) + row['1990_total']
        asia_data['2000_total'] = asia_data.get('2000_total', 0) + row['2000_total']
        asia_data['2010_total'] = asia_data.get('2010_total', 0) + row['2010_total']
    if row['Continent'] == 'Europe':
        europe_data['1980_total'] = europe_data.get('1980_total', 0) + row['1980_total']
        europe_data['1990_total'] = europe_data.get('1990_total', 0) + row['1990_total']
        europe_data['2000_total'] = europe_data.get('2000_total', 0) + row['2000_total']
        europe_data['2010_total'] = europe_data.get('2010_total', 0) + row['2010_total']
    if row['Continent'] == 'Oceania':
        oceania_data['1980_total'] = oceania_data.get('1980_total', 0) + row['1980_total']
        oceania_data['1990_total'] = oceania_data.get('1990_total', 0) + row['1990_total']
        oceania_data['2000_total'] = oceania_data.get('2000_total', 0) + row['2000_total']
        oceania_data['2010_total'] = oceania_data.get('2010_total', 0) + row['2010_total']
    if row['Continent'] == 'Latin America and the Caribbean':
        latin_data['1980_total'] = latin_data.get('1980_total', 0) + row['1980_total']
        latin_data['1990_total'] = latin_data.get('1990_total', 0) + row['1990_total']
        latin_data['2000_total'] = latin_data.get('2000_total', 0) + row['2000_total']
        latin_data['2010_total'] = latin_data.get('2010_total', 0) + row['2010_total']
    if row['Continent'] == 'Northern America':
        north_data['1980_total'] = north_data.get('1980_total', 0) + row['1980_total']
        north_data['1990_total'] = north_data.get('1990_total', 0) + row['1990_total']
        north_data['2000_total'] = north_data.get('2000_total', 0) + row['2000_total']
        north_data['2010_total'] = north_data.get('2010_total', 0) + row['2010_total']
    # data[continent] = data.get(continent
    # print(row['Continent'], row['1980_total'])
    # print(index, row)
```

```python
total_1980 = list()
total_1980.extend([asia_data.get('1980_total'), europe_data.get('1980_total'), africa_data.get('1980_total'),
                   oceania_data.get('1980_total'), latin_data.get('1980_total'), north_data.get('1980_total')])

total_1990 = list()
total_1990.extend([asia_data.get('1980_total'), europe_data.get('1980_total'), africa_data.get('1980_total'),
                   oceania_data.get('1980_total'), latin_data.get('1980_total'), north_data.get('1980_total')])

total_2000 = list()
total_2000.extend([asia_data.get('1980_total'), europe_data.get('1980_total'), africa_data.get('1980_total'),
                   oceania_data.get('1980_total'), latin_data.get('1980_total'), north_data.get('1980_total')])

total_2010 = list()
total_2010.extend([asia_data.get('1980_total'), europe_data.get('1980_total'), africa_data.get('1980_total'),
                   oceania_data.get('1980_total'), latin_data.get('1980_total'), north_data.get('1980_total')])
```

```python
df_continent_decades = pd.DataFrame({
    'Continent': df['Continent'].unique(),
    '1980_total': total_1980,
    '1990_total': total_1990,
    '2000_total': total_2010,
    '2010_total': total_2010,
    # 'Asia': asia_data.values()
})

df_continent_decades
```
