```python
def load_csv_from_google_drive(url, name):
  # Attempting to load a CSV file from a Google Drive link into a Pandas DataFrame.
  try:
      # Extracting file ID from the original URL to construct a direct download link.
      url = 'https://drive.google.com/uc?id=' + url.split('/')[-2]

      # Loading the CSV data from the constructed URL into a Pandas DataFrame.
      df = pd.read_csv(url)

      # Confirmation message upon successful loading of CSV data.
      print(f'{name} CSV loaded successfully.')

      # Returning the loaded DataFrame.
      return df

  except Exception as e:
      # Handling any exceptions that may occur during the process and printing the error message.
      print('Exception occurred: ', str(e))
```

## **Question 1**

For each team, what difference in ‘rest’ can you see, when playing ‘home’ games, v/s when playing ‘away’ games?

```python
def calculate_rest(group):
    group = group.sort_values('date')
    group['rest'] = (group['date'] - group['date'].shift()).dt.days
    return group

# Calculate rest for each team (both as home and away team)
df_home = df.groupby('Home_team').apply(calculate_rest).reset_index(drop=True)
df_away = df.groupby('Visiting_team').apply(calculate_rest).reset_index(drop=True)

# Combine home and away data
df_home['is_home'] = True
df_away['is_home'] = False
df_combined = pd.concat([df_home, df_away], ignore_index=True)

# Calculate average rest for home and away games for each team
rest_diff = df_combined.groupby(['Home_team', 'is_home'])['rest'].mean().unstack()
rest_diff.columns = ['away_rest', 'home_rest']

print(rest_diff['home_rest'] - rest_diff['away_rest'])

#Absolute Difference in rest
rest_diff['rest_difference'] = np.abs(rest_diff['home_rest'] - rest_diff['away_rest'])

# Sort by rest difference
rest_diff = rest_diff.sort_values('rest_difference', ascending=False)#.round()

rest_diff.round()
```

```python
# @title Rest difference

fig = px.line(rest_diff, x=rest_diff.index, y='rest_difference',
              title='Rest Difference',
              labels={'Home_team': 'Home Team', 'rest_difference': 'Rest Difference (Days)'})
fig.update_layout(xaxis_tickangle=45)
fig.show()
```

```python
# Create bar charts for home rest, away rest, and rest difference using Plotly Express

#Home Team Rest


# Create a bar chart for home rest
fig_home_rest = px.bar(rest_diff.sort_values(by='home_rest', ascending=False), x=rest_diff.sort_values(by='home_rest', ascending=False).index, y='home_rest',
                     labels={'Home_team': 'Team', 'home_rest': 'Average Home Rest (Days)'},
                     title='Average Home Rest for NBA Teams')
fig_home_rest.update_xaxes(tickangle=65)
fig_home_rest.show()


#Away Team Rest
fig_away_rest = px.bar(rest_diff.sort_values(by='away_rest', ascending=False), x=rest_diff.sort_values(by='away_rest', ascending=False).index, y='away_rest',
                     labels={'Home_team': 'Team', 'away_rest': 'Average Away Rest (Days)'},
                     title='Average Away Rest for NBA Teams')
fig_away_rest.update_xaxes(tickangle=65)
fig_away_rest.show()

#Team's  Difference between Rest after Home and Away game
fig_rest_diff = px.bar(rest_diff, x=rest_diff.index, y='rest_difference',
                     labels={'Home_team': 'Team', 'rest_difference': 'Absolute Rest Difference (Days)'},
                     title='Absolute Difference in Average Rest Between Home and Away Games')
fig_rest_diff.update_xaxes(tickangle=65)
fig_rest_diff.show()
```

```python
import plotly.graph_objects as go


fig = go.Figure()
fig.add_trace(go.Bar(
    x=rest_diff.index,
    y=rest_diff['home_rest'],
    name='Home Rest',
    marker_color='blue'
))
fig.add_trace(go.Bar(
    x=rest_diff.index,
    y=rest_diff['away_rest'],
    name='Away Rest',
    marker_color='red'
))
fig.add_trace(go.Bar(
    x=rest_diff.index,
    y=rest_diff['rest_difference'],
    name='Rest Difference',
    marker_color='green'
))

fig.update_layout(
    title='Home Rest, Away Rest, and Rest Difference by Team',
    xaxis_title='Team',
    yaxis_title='Average Rest Days',
    barmode='group'
)
fig.show()
```

```python
# Pivot the data to create a matrix suitable for a heatmap
heatmap_data = df_combined.pivot_table(values='rest', index='Home_team',
                                       columns='is_home', aggfunc='mean')
heatmap_data.columns = ['Away', 'Home']

# Sort in descending order y away games
heatmap_data= heatmap_data.sort_values(by=['Away'], ascending=False)


# Create the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(heatmap_data, annot=True, cmap='YlOrRd', fmt='.1f')
plt.title('Average Rest Days: Home vs Away Games')
plt.xlabel('Game Location')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

# **Question -2:  b) What is the effect that ‘rest’ has, on overall match-outcome?**

### To address this analysis, we first need to discover a way by which we could know that what is the `rest` in this statement.  

### The dataset is of matches happened in differnt weeks between different teams, there is a gap (of a week) between the next match of each team with a different teams, so in this gap which is called rest time, we need to discover whether this rest time impacts on match performance of each team or not.

## Proposed analysis solution:
### **1. First of all** we need to do one feature engineering, to make a new feature out of `date` feature called `week`. By this we get to know that which team is playing in which week of the month.
- And also we can track weekly performance of each team whether they are improving on the score or not.

### **2. The next step** is to check whether the teams are playing or eliminating in between the game, by doing so we can compare the results by validating team scores to the previous week score.
* To achieve this we will create one new features -
* **Step 1**- To extracts the week number from the date column using the `ISO calendar week`. For example, if a date falls in the first week of the year, the `game_week` value will be 1.

* **Step 2** - Create `previous_week` Column:
It uses the groupby function to group the DataFrame by the visitor_neutral_team column. This means that the data is processed separately for each team.
The shift(1) function shifts the game_week values by one position within each group. This effectively moves the week number from the current row to the previous row, aligning it with the previous game week for the same team.

* **Step 3** - Create `previous_week_game` Column:
This line creates a new column, `previous_week_game`.
It compares the `previous_week` column to the current `game_week` minus one. If the previous game was indeed in the week before the current game week, the value in `previous_week_game` will be `True`. Otherwise, it will be `False`.

#### `The previous_week` column stores the game week number of the previous game for each team, and `previous_week_game` is a boolean column indicating if the team had a game in the previous week.

### **3. Then** We need a determiner for weekly performance for each team, by which we can understand what what the score of a team before the week they are playing.

* To achieve this we will create one new features -
* **Step 1**- Group by Week and Team: Groups the data by `game_week` and `visitor_neutral_team.`

* **Step 2** - Computes the average points for each team within each week.

* **Step 3** - Reset Index converting the result back to a DataFrame with default integer indexing.

The result `weekly_performance`, contains the average points scored by each team for each week.

### **4. The next step** is the main function where we compare the results of each team and store it into dataframe, by calculating the average performance by teams in the previous week and get their difference.

* To achieve this we will create one new features -
* **Step 1**- **Merge Average Points:** Add the average points scored by each team in each week to the main DataFrame. This allows for a comparison of team performance across different weeks.

* **Step 2** - **Identify No Game in Previous Week:** Create a column that indicates whether a team did not play in the previous week. This distinction is crucial for comparing performance in weeks following rest weeks versus non-rest weeks.

* **Step 3** **- Calculate Average Points After No Game Weeks:** To calculate the average performance (points scored) of teams in weeks that follow a week with no games. This helps in understanding how rest affects team performance.

* **Step 4** **-Calculate Average Points After Game Weeks:** To calculate the average performance (points scored) of teams in weeks that follow a week with games. This serves as a comparison to the performance after rest weeks.

* **Step 5** **-Compare Performances:** To merge the two sets of performance data (after rest weeks and after non-rest weeks) and calculate the difference. This final comparison highlights whether teams perform better, worse, or the same after having a rest week.

The overall purpose of this function is to provide insights into the impact of rest on team performance. By comparing the average points scored after rest weeks versus non-rest weeks, analysts, coaches, and teams can make informed decisions about the importance of rest periods and their strategic impact on performance. This can be particularly useful for scheduling, training, and preparing teams for competitions.


### The last step is to visualize the result and the final analysis of the findings.

```python
# Renaming the feature names to appropriate one
df_2.rename(
    columns={
      'Date':'date',
      'Visitor/Neutral':'visitor_neutral_team',
      'PTS':'points',
      'Home/Neutral':'home_neutral_team',
      'PTS.1':'points_1',
      'Start (ET)':'time',
      'Unnamed: 7':'overtime_periods',
      'Unnamed: 6':'box_score',
      'Attend.':'attendance',
      'Arena':'arena'
    },
    inplace=True
)
```

## Step 1: Creating new feature called `week`

```python
# Creating week feature from date feature
df_2['date'] = pd.to_datetime(df_2['date'])
df_2['week'] = df_2['date'].dt.isocalendar().week
```

## Step 2: Tracking down which teams are playing in which weeks

```python
df_2.head(5)
```

```python
weekly_performance = df_2.groupby(['week', 'visitor_neutral_team'])['points'].mean().reset_index()
```

```python
# Sorting the values based on visitor_neutral_team and date
df_sorted = df_2.sort_values(by=['visitor_neutral_team', 'date'])
```

## Step 3: Creating weekly_performance dataframe


```python
df_sorted['game_week'] = df_sorted['date'].dt.isocalendar().week

# Create a column to identify if the team had a game in the previous week
df_sorted['previous_week'] = df_sorted.groupby('visitor_neutral_team')['game_week'].shift(1)
df_sorted['previous_week_game'] = df_sorted['previous_week'] == (df_sorted['game_week'] - 1)
```

```python
# Get the week number of each teams game
df_2['game_week'] = df_2['date'].dt.isocalendar().week

# Creating a column to identify if the team had a game in the previous week
df_2['previous_week'] = df_2.groupby('visitor_neutral_team')['game_week'].shift(1)
df_2['previous_week_game'] = df_2['previous_week'] == (df_2['game_week'] - 1)
```

## Step 4: Calculating average point and coomparing weekly *performance*

```python
weekly_performance = df_sorted.groupby(['game_week', 'visitor_neutral_team'])['points'].mean().reset_index()
```

## Step 5: Comparing the performance

```python

# Merging the average points with the previous week data
df_merged = pd.merge(df_sorted, weekly_performance, on=['game_week', 'visitor_neutral_team'], suffixes=('', '_avg'))

# Identifying weeks following a week with no games
df_merged['no_game_previous_week'] = ~df_merged['previous_week_game']

# Calculating average points in weeks following no game weeks
rest_week_performance = df_merged[df_merged['no_game_previous_week']].groupby('visitor_neutral_team')['points_avg'].mean().reset_index()

# Calculating average points in weeks with previous week games
non_rest_week_performance = df_merged[df_merged['previous_week_game']].groupby('visitor_neutral_team')['points_avg'].mean().reset_index()

# Merging both DataFrames for comparison
performance_comparison = pd.merge(rest_week_performance, non_rest_week_performance, on='visitor_neutral_team', suffixes=('_after_rest', '_after_non_rest'))
performance_comparison['difference'] = performance_comparison['points_avg_after_rest'] - performance_comparison['points_avg_after_non_rest']
```

```python
performance_comparison['difference'].sum()
```

## Plotting the result

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))
barplot = sns.barplot(x='visitor_neutral_team', y='difference', data=performance_comparison, palette='viridis')


# Add title and labels
plt.title('Difference in Average Points After Rest Week vs Non-Rest Week')
plt.xlabel('Team')
plt.ylabel('Difference in Average Points')
plt.xticks(rotation=90)

for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.1f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'center',
                     xytext = (0, 9),
                     textcoords = 'offset points')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show plot
plt.show()
```

# As the above graph represents numerical difference in the average points scored by the team after a rest week compared to a non-rest week.

### **Positive bars:**  We can see the mean line which is 0.0 the teams having their bar above the 0.0 values witnessed a increase in average performance by showen number averagly compared to a non-rest week.

### **Negative bars:** If a bar is below the zero line, it means the visiting team scored fewer points on average after a rest week than after a non-rest week.

### The team performance are not much affected by the rest week.

#### Only some teams such as `Dallas Mavricks`, `Indiana Pacers`, `Milwaukee Bucks`, `Phoenix Suns` and `Utah Jazz` witnessed a drasting change in their scores.

## Question 3: When are teams more victorious- when playing at home, or when playing away?

```python
df = pd.concat([df_october, df_november, df_december]) #merge all three datasets for months October, November and December
```

```python
# renaming the columns
df.rename(columns={
    "Start (ET)":"Start_time",
    "Visitor/Neutral":"Visiting_team",
    "PTS":"Visiting_team_PTS",
    "Home/Neutral":"Home_team",
    "PTS.1":"Home_team_PTS",
    "Unnamed: 6":"IsBoxScore",
    "Unnamed: 7":"Overtime",
    "Attend.":"Attendance",
},inplace=True)
```

```python

#creating a new column in dataset contains the winning team by comparing the visiting and home team points
df['Winner'] = np.where(df['Visiting_team_PTS'] > df['Home_team_PTS'],df["Visiting_team"],df['Home_team'])
```

```python
#creating a dictionary containing visting team victories and home team victories
dict={}
dict["visit_vict"]=0
dict["home_vict"]=0
for i in range(df.shape[0]):
    if(df.iloc[i, 2]==df.iloc[i,11]):
        dict["visit_vict"]+=1
    else:
        dict["home_vict"]+=1
```

```python

place = list(dict.keys())
values = list(dict.values())
plots=plt.bar(place, values, color ='green', width = 0.4)
#getting center position of the bars to annotate the wins
for bar in plots.patches:
    x_pos = bar.get_x() + bar.get_width() / 2
    y_pos = bar.get_height()
    plt.annotate(format(y_pos), (x_pos, y_pos), ha='center', va='center', size=10, xytext=(0, 8), textcoords='offset points')
plt.xlabel("Match held at")
plt.ylabel("No of Wins")
plt.title("No of matches won by teams at home and visiting")
plt.show()
```

### The above Bar plot clearly demonstrates the effect of the location, i.e., ground, on the result of the match, with 291 victories coming when played at home ground and 198 when played on away grounds
### this may be atrributed to the fact that the teams practice more on thier home grounds and recieve extra support and cheers from home croud to boost thier morale

```python
#adding the columns with true or false if visting team wins or loses and same for home team
df['Visiting_team_wins'] = df['Visiting_team_PTS'] > df['Home_team_PTS']
df['Home_team_wins'] = df['Visiting_team_PTS'] < df['Home_team_PTS']

#taking the unique teams from visiting and home
teams = pd.unique(df[['Visiting_team', 'Home_team']].values.ravel('K'))

#calculating win percentages of every team
win_percentages = {}

for team in teams:
    visiting_wins = df[df['Visiting_team'] == team]['Visiting_team_wins'].sum()
    home_wins = df[df['Home_team'] == team]['Home_team_wins'].sum()

    total_games = visiting_wins + home_wins
    if total_games > 0:
        visiting_win_percentage = (visiting_wins / total_games) * 100
        home_win_percentage = (home_wins / total_games) * 100
        win_percentages[team] = (visiting_win_percentage, home_win_percentage)


num_teams = len(teams)
num_cols = 5
num_rows = 6
#creating sub plots for all the teams
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, 25))
axes = axes.flatten()


for ax, (team, (visiting_percentage, home_percentage)) in zip(axes, win_percentages.items()):
    ax.pie([visiting_percentage, home_percentage], labels=['Visiting Wins', 'Home Wins'], autopct='%1.1f%%', colors=['#ff9999','#66b3ff'])
    ax.set_title(f'{team}')


for ax in axes[len(teams):]:
    ax.axis('off')

plt.tight_layout()
plt.show()
```

### A similar trend is continued even when seen at the individual team level, where most of the teams perform well in the home ground compared to the away ground. only a few teams have more away victories; the reason may be unyielding conditions at home when compared to the other locations.
